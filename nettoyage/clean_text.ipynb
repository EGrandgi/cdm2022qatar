```python
import json
from unidecode import unidecode
import nltk
from stop_words import get_stop_words
path = '../scraping/json/'
filename = '2019-02-20_all_articles_courrierint.json'
f = open(path + filename, encoding='utf-8')
doc = json.load(f)
```

```python
doc['content'].values()
```

```python
def sentense2cleanTokens(sent):
    newsent=""
    stopw=get_stop_words('fr')
    words = sent.split() 
    for r in words: 
        if r not in stopw: 
            newsent += " " +r
    #sent = unidecode(sent)
    newsent = newsent.replace ("\n", "")
    #sent = sent.replace ('\', "")
    newsent = newsent.replace("\xad", "")
    #sent = sent.lower()
    #sent = "".join([x if x.isalnum() else " " for x in sent])
    newsent = " ".join(newsent.split())
    return newsent

content = doc['content'].values()
#print(content)
cleancontent=[sentense2cleanTokens(x) for x in content]
cleancontent
```
